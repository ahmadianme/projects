```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
def adalineNeuron(X, W, b):
    net = np.sum(np.multiply(X[0], W)) + b
    if net >= 0:
        return 1, net
    else:
        return -1, net
```

```python
def generateSamples():
    class1 = []
    for i in range(100):
        class1.append(([np.random.normal(2, 0.5), np.random.normal(0, 0.2)], 1))

    class2 = []
    for i in range(30):
        class2.append(([np.random.normal(0, 0.1), np.random.normal(1, 0.7)], -1))
        
    data = class1 + class2
    random.shuffle(data)
        
    return data, class1, class2
```

```python
def train(data):
    W = [np.random.uniform(-0.005, .005), np.random.uniform(-0.005, .005)]
    b = np.random.uniform(-0.005, .005)
    lr = 0.0001
    endThreshold = 1

    costs = []
    
    epoch = 1
    for ep in range(1000):
        cost = 0
        for datum in data:
            h, net = adalineNeuron(datum, W, b)

            error = datum[1] - net
            cost += pow(error, 2) / 2

            for i in range(len(W)):
                W[i] = W[i] + lr * (error) * datum[0][i]

            b = b + lr * (error)
            
        costs.append(cost)
        
        print('Epoch ' + str(epoch) + ': Cost: ' + str(cost))
    
        if cost < endThreshold:
            break;
                
        epoch += 1
        
    return W, b, costs
```

```python
samples, class1, class2 = generateSamples()
W, b, costs = train(samples)

print()
print('Final Weights: ' + str(W))
print('Final Bias: ' + str(b))
```
